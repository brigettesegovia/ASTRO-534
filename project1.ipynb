# BACKGROUND

Big Bang Nucleosynthesis (BBN) predicts the abundance of ${}^4\text{He}$. Once temperatures reached $\sim 10^9 K$, ${}^4\text{He}$ was immediately developed (after the formation of deuterium). ${}^4\text{He}$ is a molecule that is stable enough to resist significant decay. It is an important probe for the baryon-to-photon ratio $\eta$ in the early universe as it directly relates to this ratio. Up to now, stellar nucleosynthesis alone has failed to explain the abundance of ${}^4\text{He}$ found in the universe, making ${}^4\text{He}$ an interestig probe for BBN conditions. 

# FREEZE-OUT OF STABLE PARTICLES (main reference 3.3.5 MBW)

Consider a species  $i$  taking part in the following two body interactions:

$i + j \leftrightarrow a + b$.  


The production and destruction rates of $i$ due to the interaction rate will be $\alpha (T)$ and $\beta (T)$. The equation that describes how particles are destroyed due to interaction with $j$ and created due to the interactions between $a + b$ is given by:   

\begin{equation*}
\frac{dn_i}{dt} + 3H(t)n_i = \alpha (T) n_{\alpha} n_{b} + \beta (T) n_i n_j . \tag{1}  
\end{equation*}

Where $\alpha (T) n_{\alpha} n_{b}$ refers to the production rate of particles, and $\beta (T) n_i n_j$ refers to the destruction rate of particles. Note that using this same logic, you can obtain a similar equation for $j$.  

Assuming $a$ and $b$ are generally in equilibrium with the general hot plasma, with $T_a = T_b = T$ (temperatures), while $i$ and $j$ are coupled to their reactions with $a$ and $b$, we define an equilibrium density for $i \rightarrow n_{i,eq}$ and $j \rightarrow n_{j,eq}$.  

The destruction rate then can be written as: $\beta (T) n_{i,eq}n_{j,eq} = \alpha (T)n_a n_b$. Where once again the $n_{i,eq}$ and $n_{j,eq}$ mentioned here are the number densities of each particle under the assumption of thermal equilibrium with the hot plasma.   

Considering the case that $j$ and $b$ are the antiparticles of $i$ and $a$, as long as the chemical potential of $i$ is small, we can say $n_i = n_j$.  

The rate equation then becomes:  

\begin{equation*}
\frac{dn_i}{dt} + 3H(t)n_i = \beta (T) (n_{i, eq}^2 - n_{i}^2).
\end{equation*}

In the case for neutron freeze-out, the interactions are dominated by two main cases. 

| Case 1:            | Case 2:            |
|-----------------|-----------------|
$n+\nu \leftrightarrow p+e^-$   |  $n+e^+ \leftrightarrow p+\overline \nu$       |
    

Going back to the interaction rate equation above, $\textbf{CASE 1}$ can be rewritten as: 

\begin{equation*}
\frac{dn_n}{dt} + 3H(t)n_n = \alpha (T)n_pn_{e^-}  + \beta(T)n_nn_{\nu}. \tag{2}
\end{equation*} 


Where $ \alpha (T)n_pn_{e^-}$ refers to the creation rate and $\beta(T)n_nn_{\nu}$ refers to the destruction rate.   

The right hand side equals 0 when we assume a state of equilibrium. The justification is as follows: 

$\Rightarrow \alpha (T)n_{p,eq}n_{e^-} = \beta (T)n_{n,eq}n_{\nu} \Rightarrow \alpha (T)n_{p,eq}n_{e^-} - \beta (T)n_{n,eq}n_{\nu} = 0$  

  

Now combining both equations 1 and 2 (similar process to what MBW does in equation 3.164), we get:  
\begin{equation*}
\frac{dn_n}{dt} + 3H(t)n_n = \alpha (T) [n_pn_{e^-} - n_{p,eq}n_{e^-}] - \beta (T) [n_nn_{\nu} - n_{n,eq}n_{\nu}] \tag{3}
\end{equation*}

\begin{equation*}
\frac{1}{n_b} \frac{dn_n}{dt} + 3H(t)n_n = \alpha (T) \frac{n_{e^-}}{n_b}[n_p - n_{p,eq}] - \beta (T) \frac{n_{\nu}}{n_b} [n_n - n_{n,eq}] \tag{4}
\end{equation*}

Where I have divided by $n_b$ (the number density of baryons) on both sides, and using a clear substitution of $X_n = \frac{n_n}{n_b}$. 
Which implies that $\frac{dX_n}{dt} = \frac{1}{n_b} \frac{dn_n}{dt} - \frac{n_n}{n_b^2}\frac{dn_b}{dt}$.   
And using the assumption that baryon number is conserved: 

\begin{equation*}
\frac{dn_b}{dt} + 3H(t)n_b = 0 
\end{equation*}

\begin{equation*}
\Rightarrow -3H(t)n_b = \frac{dn_b}{dt}  
\end{equation*}

\begin{equation*}
\Rightarrow \frac{-1}{n_b} \frac{dn_n}{dt} + \frac{dX_n}{dt} = \frac {n_n}{n_b}3H(t).
\end{equation*}  

Equation (4) simplifies to:
\begin{equation*}
\frac{dX_n}{dt} = \alpha (T) \frac{n_{e^-}}{n_b} [n_p - n_{p,eq}] - \beta (T) \frac{n_{\nu}}{n_b} [n_n - n_{n,eq}].
\end{equation*}  

Distributing the $\frac{1}{n_b}$ and using $X_i = \frac{n_i}{n_b}$, you get: 

\begin{equation*}
\frac{dX_n}{dt} = \alpha (T) n_{e^-} [X_p - X_{p,eq}] - \beta (T) n_{\nu} [X_n - X_{n,eq}]. \tag{5}
\end{equation*}  

Due to the symmetry of the problem, the calculation for $\textbf{CASE 2}$ would be fairly similar, but the coefficients would end up being multiplied by a different number density. The result would be as follows: 

\begin{equation*}
\frac {dX_n}{dt} = [\alpha (T) n_{e^-} + \alpha(T)_2 n_p][X_p - X_{p,eq}] - [\beta (T)n_{\nu} + \beta (T)_2 n_{e^-}][X_n - X_{n,eq}]. \tag{6}
\end{equation*}  

Here the $[\alpha (T) n_{e^-} + \alpha(T)_2 n_p]$ accounts for the reaction $protons \rightarrow neutrons$ and $[\beta (T)n_{\nu} + \beta (T)_2 n_{e^-}]$ account for $neutrons \rightarrow protons$.   

Note that $\alpha(T)$ is different from $\alpha (T)_2$ (same thing for $\beta (T)$ and $\beta (T)_2$).  

Simplifying some notation: 

\begin{equation*}
\frac{dX_n}{dt} = (p \rightarrow n) [X_p - X_{p,eq}] - (n \rightarrow p)[X_n - X_{n,eq}]
\end{equation*}

\begin{equation*}
\frac{dX_n}{dt} = (p \rightarrow n) X_p - (p \rightarrow n) X_{p,eq} - (n \rightarrow p)X_n - (n \rightarrow p)X_{n,eq}
\end{equation*}  

saying $X_p + X_n = 1$ since the mass fraction / abundance of neutrons and protons make up the total of the baryons: 
\begin{equation*}
(p \rightarrow n) [1 - X_n] - (p \rightarrow n)[1 -  X_{n,eq}] - (n \rightarrow p)X_n - (n \rightarrow p)X_{n,eq}
\end{equation*}  

\begin{equation*}
\Rightarrow [X_{n,eq} - X_n](p \rightarrow n) - [X_{n,eq} - X_n](n \rightarrow p)
\end{equation*} 

\begin{equation*}
\Rightarrow \frac{dX_n}{dt} = -[(p \rightarrow n) + (n \rightarrow p) ] [X_n + X_{n,eq}].   \tag{7}
\end{equation*} 

Equation (7) is now the complete reaction rate for neutrons. This equation can be found in [Mukhanov 2003](https://arxiv.org/abs/astro-ph/0303073)).

# SOLVING FOR TEMPERATURE 
### (two ways shown: (1) simple way using lecture material, (2) harder way using derivations; both require assumptions)

An $\textbf{easy way}$ to find temperature for this system could be done as follows (all values used here are from lecture):  

Once again, we are assuming that neutrinos and positrons are in thermal equilibrium with the plasma, hence $T_{\nu} = T_{e^-} = T$ and that $Q = (m_n - m_p)c^2 \approx 1.3 MeV$ (this is differene in kinetic energy between the neutrons and protons).   

We can add to our assumptions by claiming that $\eta = \frac{n_b}{n_{\gamma}} = 6 \times 10^{-10}$ which corresponds to $\frac{n_n}{n_p} \approx 0.16$.  

In lecture we learned that the equation that relates $Q$ with neutron to proton ration is: 

\begin{equation*}
\frac{n_n}{n_p} = e ^{- \frac{Q}{kT}} \tag{8}
\end{equation*}

\begin{equation*}
\Rightarrow ln(\frac{n_n}{n_p}) = - \frac{Q}{KT} \Rightarrow kT = \frac{-Q}{ln(\frac{n_n}{n_p})} = -ln(0.16) \times 1.3 MeV = 0.71 MeV. 
\end{equation*}


A $\textbf{more in depth}$ way to solve for this would be using the interaction rate equation from below which governs this set up: 

\begin{equation*}
\lambda_{ne\rightarrow p\nu} \simeq \lambda_{n\nu \rightarrow pe}(y) \simeq \frac{1+3g^2_A}{2\pi^3} \frac{G_F^2}{\hbar} \left(\frac{Q}{y}\right)^5 \left[\frac{45 \zeta (5)}{2} + \frac{7 \pi ^4 y}{60} + \frac{3\zeta (3)}{2} \left(1 - \frac{1}{2} \left(\frac{m_e c^2}{Q}\right)^2\right) y^2\right] \tag{9}
\end{equation*}
This is equation (8) from [Ukhanov 2003](https://arxiv.org/pdf/astro-ph/0303073).

for $\lambda_{p\nu\rightarrow ne}$ & $\lambda_{pe\rightarrow n\nu}$ $ \simeq e^{-y} \lambda_{n\nu\rightarrow pe}(y) $

\begin{equation*}
\Rightarrow \frac{dX_n}{dt}\simeq -2(1+e^{-y}) \lambda_{n\nu \rightarrow pe}(y)(X_n - X_{n,eq}) \tag{10}
\end{equation*}

To solve this, we can convert to y space:   

\begin{equation*}
\frac{dX_n}{dt} = \frac{dT}{dt} \frac{dy}{dT} \frac{dX_n}{dy} = \left(\frac{4\pi^3}{45}g_*\right)^{1/2} \left(\frac{G}{c^5 \hbar^3}\right)^{1/2} \frac{Q^2}{y} \frac{dX_n}{dy} \tag{11}
\end{equation*}


Now equation (10) becomes (by combining it with equation (11)): 


\begin{equation*}
\frac{dX_n}{dy} \simeq -2\left(1+e^{-y}\right) \left( \left(\frac{4\pi^3}{45} g_* \right)^{-1/2} \left(\frac{G}{c^5 \hbar^3} \right)^{-1/2} \frac{y}{Q^2} \right) 
\times \left( \frac{1+3g_A^2}{2\pi^3} \frac{G_F^2}{\hbar} \left(\frac{Q}{y}\right)^5 \left[\frac{45 \zeta(5)}{2} + \frac{7 \pi^4 y}{60} + \frac{3\zeta(3)}{2} \left(1 - \frac{1}{2} \left(\frac{m_e c^2}{Q}\right)^2 \right) y^2 \right] \right) 
\times \left(X_n - X_{n,eq} \right)
\end{equation*}


\begin{equation*}
\Rightarrow \frac{dX_n}{dy} \simeq -2.858 \, \frac{\left(1+e^{-y}\right) {\,1+0.487y+0.071y^2}}{y^4} \left( X_n - \frac{1}{1+e^y} \right)  \tag{12}
\end{equation*}
Where we used the fact that $X_{n,eq} = \frac{1}{1+e^y}$

after plugging in for constants and known values. One final thing to consider is the change of this equation when accounting for neutron decay. With neutron decay, the full form is:
  
\begin{equation*}
 \frac{dX_n}{dy} \simeq -2.858 \, \frac{\left(1+e^{-y}\right) {\,1+0.487y+0.071y^2}}{y^4} \left( X_n - \frac{1}{1+e^y} \right) - 10^{-3}yX_n   \tag{13}
\end{equation*} 
where $10^{-3}y$ corresponds to a factor of $1/\tau$ where $\tau = 886s$ (the lifetime of a neutron).   

Below I show how you can integrate and get the value of X_n for freeze out, then get the value of y, and lastly, the temperature at the time of freeze-out.


```python
# Importing the necessary libraries

import numpy as np
from astropy import units as u
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from astropy import constants as const

# set the default plotting parameters
plt.rcParams.update({
    'savefig.bbox': 'tight',
    'savefig.dpi': 200,
    'figure.dpi': 200,
    'lines.linewidth': 1.5,
    'font.family': 'sans-serif',
    'font.sans-serif': 'Verdana',
    'font.size': 10,
    'axes.labelsize': 15,
    'xtick.direction': 'in',
    'ytick.direction': 'in',
    'xtick.top': True,
    'ytick.right': True,
    'xtick.major.pad': 5,
    'xtick.minor.pad': 5,
    'ytick.major.pad': 5,
    'ytick.minor.pad': 5,
    'xtick.major.size': 5,
    'ytick.major.size': 5,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'legend.fontsize': 12,
    'legend.frameon': False,
    'text.usetex': True
})
```

```python
def dX_dy(y, X):
    """Differential equation for the neutron abundance X as a function of y."""
    if y < 1e-6:  
        return 0
    factor = -2.858 * (1 + np.exp(-y)) * (1 + 0.487*y + 0.071*y**2) / max(y**4, 1e-6)
    return factor * (X - 1/(1 + np.exp(y))) - 1e-3 * y * X


# Integral range
y_min = 1e-3  # Start of integration
y_max = 10    # Approximate y -> infinity
X0 = [1/(1 + np.exp(y_min))]  # Initial neutron abundance from equilibrium

# Solve the ODE
sol = solve_ivp(dX_dy, [y_min, y_max], X0, method='LSODA', t_eval=np.logspace(np.log10(y_min), np.log10(y_max), 100))


freeze_out_value = sol.y[0][-1]  # Correct indexing for the last value
print(f'Freeze-out abundance X_n(y -> infinity) ≃{freeze_out_value:.3f}')
```

```python
# y is simply calculated from X
y =  np.log(1/freeze_out_value -1) 
print(f'Corresponding y value ≃{y:.3f}')
```

```python
# T is given by the relation T = 1.3 MeV / y
T = 1.3*u.MeV / y
print(f'Temperature at freeze-out ≃{T:.3f}')
```

```python
# Define the rate equation for X_n as a function of y.
def rate_equation(y, X):
    """The rate equation without the neutron decay term."""
    return -2.858 * (1 + np.exp(-y)) * (1 + 0.487 * y + 0.071 * y**2) * (X - 1/(1 + np.exp(y))) / y**4


# Create an array of y values.
y_vals = 10**np.linspace(-6, 2, 101)

# The equilibrium solution 
Xn_eq = 1 / (1 + np.exp(y_vals))

# Rate Equation
sol = solve_ivp(rate_equation, [y_vals[0], y_vals[-1]], [0.25], t_eval=y_vals, method='LSODA')
X_n = sol.y[0]

# Rate equation with neutron decay
sol_decay = solve_ivp(dX_dy, [y_vals[0], y_vals[-1]], [0.25], t_eval=y_vals, method='LSODA')
X_n_neutron_decay = sol_decay.y[0]




# Making a plot
fig, ax = plt.subplots()

# Plot each solution on a log-log scale.
ax.plot(np.log10(y_vals), np.log10(X_n), lw=1, c='midnightblue', label='$X_n$')
ax.plot(np.log10(y_vals), np.log10(X_n_neutron_decay), lw=1, c='plum', ls=':', label='$X_{n, decay}$')
ax.plot(np.log10(y_vals), np.log10(Xn_eq), lw=1, c='red', ls='--', label='$X_{n, eq}$')

# Label axes and set tick parameters.
ax.set_xlabel(r'$\log_{10}y$')
ax.set_ylabel(r'$\log_{10}X$')
ax.set_xlim(-1, 1.5)
ax.set_xticks([-1, -0.5, 0, 0.5, 1, 1.5])
ax.set_ylim(-1, -0.2)
ax.set_yticks([-1, -0.8, -0.6, -0.4, -0.2])
ax.legend()

plt.show()
```

# ABUNDANCE OF HELIUM 4

In order to solve for the ${}^4He$ abundance as a function of $\eta$, we need to have the time at which deuterium formation stopped working such that ${}^4He$ could start being formed. From MBW 3.210:

\begin{equation*}
X_D = 16.4X_nX_p\eta (\frac{m_N}{T})^{-3/2} exp(\frac{2.22MeV}{T})  \tag{14}
\end{equation*}

Rewriting this in terms of temperature:

\begin{equation*}
2.22MeV ( ln[\frac{X_D (\frac{m_N}{T})^{3/2} }{16.4X_nX_p \eta} ])^{-1} = k_BT 
\end{equation*}

Note that below I show how $ log(\frac{m_N}{T}^{3/2}) \approx 13.95$.

Now the equation becomes:

\begin{equation*}
\Rightarrow k_BT = \frac{B_D}{ln(X_D/\eta) + 13.95}
\end{equation*}

The equilibrium deuterium abundance equation in Big Bang Nucleosynthesis (BBN) is typically represented as: $X_D, c \approx 1.5 \times 10^{-14} /\eta$ (this value can be found in [Mukhanov 2003](https://arxiv.org/abs/astro-ph/0303073))
Using this in our equation above we get:
\begin{equation*}
\Rightarrow k_BT = \frac{B_D}{ln(1.5 \times 10^{-14} /\eta^2) + 13.95}
\end{equation*}  

and using the radiation dominated equation to get time: 
\begin{equation*}
\Rightarrow t_D = 1.4(1MeV/k_BT)
\end{equation*}  


$t_D \approx 1.4 \cdot \left[ 2/2.22 \ln \left( \frac{1}{\eta} \right) - 17.45/2.22 \right]^2 \, \text{s}$.   

Which leads us to 

\begin{equation}
X_{nD} = 0.157 \times exp(-t_D/886)
\end{equation}

Which lastly gives us: 

\begin{equation}
X_4He = 2 \times X_{nD}
\end{equation}
As given by MBW. 

Now we can visualize this relationship.

```python
# Calculate the term (m_N * c^2) / (k_B * T)
T = 1e9 * u.K  # guess in temperature during deuterium bottle neck (if you do a google search it is ~1e9)
term = (const.m_n * const.c**2) / (const.k_B * T)
 
 
# Calculate (3/2) * ln(term)
result = (3/2) *  np.log(term)
print(f"The result is approximately: {result:.2f}")
```

```python
# compute 4He abundance
eta = 10**np.linspace(-11, -8, 100+1)

t_D = 1.4 * (2/2.22 *   np.log(1/eta) -17.45/2.22)**2 

X_nD = 0.157 * np.exp(-t_D/886)
X_4He = 2 * X_nD
```

```python
# make plot
fig, ax = plt.subplots()

ax.plot(np.log10(eta), X_4He, c='k', lw=1, label=r'$X_{\rm ^4He,now}$')

ax.fill_between([np.log10(6.05e-10),np.log10(6.37e-10)], [0,0], [1,1], 
                facecolor='purple', edgecolor=None, alpha=0.2, label= 'WMAP 2008')

ax.fill_between([-12, -7], [0.232,0.232], [0.258,0.258], 
                facecolor='plum', edgecolor=None, alpha=0.2, label=r'Coc\&Vangioni 2010')


ax.set_xlabel(r'$\log_{10} \eta$')
ax.set_ylabel(r'$\log_{10}X$')

ax.set_xlim(-11,-8)
ax.set_ylim(0.2, 0.3)

ax.legend()
plt.show()
```

Our approximation was very good because it lies on the expected values found by [WMAP](https://iopscience.iop.org/article/10.1088/0067-0049/180/2/306) and [Coc & Vangioni](https://iopscience-iop-org.proxy.lib.umich.edu/article/10.1088/1742-6596/202/1/012001).
